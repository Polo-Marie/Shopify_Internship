```python
import pandas as pd
```

```python
# Files to load
file_to_load = "2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv"

# Read the CSV into a data frame.
sneakers_df = pd.read_csv(file_to_load)
sneakers_df.head(20)
```

```python
# Get number of total orders placed in dataframe
sneakers_df.count()
```

```python
# Calculate total order_amount
order_amount = sneakers_df["order_amount"].sum()
order_amount
```

```python
# Calculate/ confirm average order value (AOC) = Revenue / total orders placed
AOC = order_amount/5000
AOC
```

```python
# Get number of total items with a "total_items" number over 10

High_order = sneakers_df.loc[(sneakers_df["total_items"] > 10)]
High_order
```

```python
High_order.count()
```

```python
new_sneakers_df = sneakers_df.copy()
new_sneakers_df.head(20)
```

```python
# Remove high items from dataframe
# https://www.geeksforgeeks.org/drop-rows-from-the-dataframe-based-on-certain-condition-applied-on-a-column/?ref=lbp
new_sneakers_df.drop(new_sneakers_df[new_sneakers_df['total_items'] > 20].index, inplace = True)
```

```python
new_sneakers_df.head(20)
```

```python
new_sneakers_df.count()
```

```python
# Calculate NEW total order_amount
order_amount = new_sneakers_df["order_amount"].sum()
order_amount
```

```python
# Calculate/ confirm NEW average order value (AOC) = Revenue / total orders placed
AOC = order_amount/4983
AOC
```
